```python
import numpy as np
import matplotlib.pyplot as plt

def read_data_from_file(filename):
    # Inizializza due liste vuote per memorizzare i dati
    pi = []
    err = []
    
    # Apre il file e legge i dati riga per riga
    with open(filename, 'r') as file:  # Modalità "r" per la lettura
        for line in file:
            # Divide ogni riga in due numeri e li aggiunge alle rispettive liste
            values = line.split()
            pi.append(float(values[0]))
            err.append(float(values[1]))
            
    return pi, err

# Codice per generare il grafico dopo aver letto i dati dal file
pi, err = read_data_from_file('pi.dat')

plt.hist(pi, bins=50)  # Crea un istogramma delle medie
plt.xlabel('Valore di pi')  # Aggiunge un'etichetta sull'asse x
plt.xlim(3.025, 3.16)  # Imposta i limiti dell'asse x
plt.ylabel('Frequenza')  # Aggiunge un'etichetta sull'asse y
plt.title("Valore di pi")  # Aggiunge un titolo al sottoplot

plt.tight_layout()  # Migliora la disposizione dei sottoplot nella figura

plt.show()  # Visualizza i grafici
```

```python
import numpy as np
import matplotlib.pyplot as plt

def read_data_from_file(filename):
    # Inizializza due liste vuote per memorizzare i dati
    pi = []
    err = []
    
    # Apre il file e legge i dati riga per riga
    with open(filename, 'r') as file:  # Modalità "r" per la lettura
        for line in file:
            # Divide ogni riga in due numeri e li aggiunge alle rispettive liste
            values = line.split()
            pi.append(float(values[0]))
            err.append(float(values[1]))
            
    return pi, err

# Codice per generare il grafico dopo aver letto i dati dal file
pi, err = read_data_from_file('pi.dat')# Definizione degli indici (x)
x = np.arange(len(pi), dtype=float)
x *= (100000/len(pi))
plt.errorbar(x, pi, yerr=err)

# Creazione del grafico con valori di pi e barre d'errore
plt.title('Valore di pi')
plt.xlabel('Numero di lanci')
plt.ylabel('Valore di pi')
plt.grid(True)
plt.show()
```
