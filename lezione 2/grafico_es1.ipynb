```python
import numpy as np
import matplotlib.pyplot as plt

def read_data_from_file(filename):
    # Inizializza due liste vuote per memorizzare i dati
    integr = []
    integr_err = []
    integr_hom = []
    err_hom = []

    # Apre il file e legge i dati riga per riga
    with open(filename, 'r') as file:  # Modalità "r" per la lettura
        for line in file:
            # Divide ogni riga in due numeri e li aggiunge alle rispettive liste
            values = line.split()
            integr.append(float(values[0]))
            integr_err.append(float(values[1]))
            integr_hom.append(float(values[2]))
            err_hom.append(float(values[3]))
            
    return integr, integr_err, integr_hom, err_hom

# Chiamata alla funzione per leggere i dati dal file
integr, integr_err, integr_hom, err_hom = read_data_from_file("risultati.dat")

x = np.arange(100, dtype=float)  # Specifica il dtype come float
x *= (10000/100)  # Number of throws = block * (Number of throws in each block)
plt.errorbar(x, np.array(integr), yerr=np.array(integr_err))
plt.xlabel('Numero di lanci')
plt.ylabel('Valore integrale')
plt.grid(True)
plt.show()

plt.errorbar(x,np.array(integr_hom),yerr=np.array(err_hom))
plt.xlabel('Numero di lanci')
plt.ylabel('Valore integrale hom')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def read_data_from_file(filename):
    # Inizializza due liste vuote per memorizzare i dati
    integr = []
    integr_err = []
    integr_hom = []
    err_hom = []

    # Apre il file e legge i dati riga per riga
    with open(filename, 'r') as file:  # Modalità "r" per la lettura
        for line in file:
            # Divide ogni riga in due numeri e li aggiunge alle rispettive liste
            values = line.split()
            integr.append(float(values[0]))
            integr_err.append(float(values[1]))
            integr_hom.append(float(values[2]))
            err_hom.append(float(values[3]))
            
    return integr, integr_err, integr_hom, err_hom

# Chiamata alla funzione per leggere i dati dal file
integr, integr_err, integr_hom, err_hom = read_data_from_file("risultati.dat")

x = np.arange(100, dtype=float)  # Specifica il dtype come float
x *= (10000/100)  # Number of throws = block * (Number of throws in each block)

# Creazione del grafico sovrapposto
plt.errorbar(x, np.array(integr), yerr=np.array(integr_err), label='Valore integrale', color='blue')
plt.errorbar(x, np.array(integr_hom), yerr=np.array(err_hom), label='Valore integrale hom', color='red')
plt.xlabel('Numero di lanci')
plt.ylabel('Valore integrale')
plt.legend()
plt.grid(True)
plt.show()
```
